```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import freqz

# Sampling frequency
fs = 44100  # 44.1 kHz

# LPF filter coefficients (scaled by 32768)
coeffs_lpf = [
    -236, -589, -390, -574, -457, -334, -69, 236, 554, 797, 901, 806, 494, -12,
    -634, -1254, -1723, -1893, -1637, -885, 362, 2019, 3926, 5868, 7606, 8913,
    9615, 9615, 8913, 7606, 5868, 3926, 2019, 362, -885, -1637, -1893, -1723,
    -1254, -634, -12, 494, 806, 901, 797, 554, 236, -69, -334, -457, -574, -390,
    -589, -236
]

# Scale coefficients (divide by 32768 as they are in fixed-point format)
coeffs_lpf = np.array(coeffs_lpf) / 32768

# Compute frequency response
w, h_lpf = freqz(coeffs_lpf, worN=2048, fs=fs)

# Magnitude in dB
mag_lpf = 20 * np.log10(np.abs(h_lpf))

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(w, mag_lpf, label='LPF (0-2700 Hz)')
plt.title('Magnitude Response of Low-Pass Filter')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude (dB)')
plt.grid(True)
plt.legend()
plt.xlim(0, fs/2)
plt.tight_layout()
plt.show()
```
