```python
import numpy as np
from scipy.signal import firwin, freqz
import matplotlib.pyplot as plt

# Design parameters
num_taps = 64
cutoff = 3000  # Hz
fs = 48000  # Hz
window = 'hamming'

# Generate LPF coefficients
lpf_coeffs = firwin(num_taps, cutoff, fs=fs, window=window)

# Frequencies to evaluate (matching test frequencies)
freqs = [1000, 1500, 3000, 5000, 23500]

# Compute frequency response
w, h = freqz(lpf_coeffs, worN=2048, fs=fs)

# Convert to magnitude (dB) and phase (degrees)
mag_db = 20 * np.log10(np.abs(h))
phase_deg = np.angle(h, deg=True)

# Print results
print("LPF Frequency Response Analysis")
print(f"Sampling Frequency: {fs} Hz")
print(f"Number of Taps: {num_taps}")
print(f"Coefficient Sum: {np.sum(lpf_coeffs):.6f} (Expected ~1 for 0 dB passband gain)")
print(f"\nCoefficients (for firmware, float32_t lpf_coeffs[{num_taps}]):")
print("{")
for i, coeff in enumerate(lpf_coeffs):
    print(f"    {coeff:.6f}f," + (" // " + str(i) if i % 8 == 7 else ""), end="\n")
print("};")
print("\nFrequency (Hz) | Magnitude (dB) | Phase (deg)")
print("-" * 40)
for f in freqs:
    idx = np.argmin(np.abs(w - f))
    print(f"{f:14.0f} | {mag_db[idx]:14.6f} | {phase_deg[idx]:14.6f}")

# Optional: Plot for visual confirmation
plt.figure(figsize=(8, 5))
plt.plot(w, mag_db)
plt.title("LPF Magnitude Response")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude (dB)")
plt.grid(True)
plt.xlim(0, 8000)
plt.axvline(cutoff, color='r', linestyle='--', label=f'{cutoff} Hz Cutoff')
plt.legend()
plt.show()
```
